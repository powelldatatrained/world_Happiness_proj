# Using Zscore to remove Outliers

```python
# importing zscore library
from scipy.stats import zscore
```

```python
# removing outliers 
outliers_features= np.abs(zscore(WHS))
threshold=3
print('Outliers:- \n',np.where(outliers_features>3))
```

These are the index position where they are outliers presence. Printing two indexs position samples to confirm if truly  threshold is >3

```python
# z score for index position 27,6 and 40,1
z_score = outliers_features
print('zscore for index position 27 and 6:', z_score.iloc[27,6])
print('zscore for index position 40 and 1:', z_score.iloc[40,1])
```

The above szcore position indexes samples shows they are >3. Since this has been confirmed, I will eliminate the positions where they are outliers

```python
# Removing outliers and storing data in new variable
new_whs = WHS[(z_score <3).all(axis =1)]
print('Before removing outliers:', WHS.shape)
print('After removing outliers:', new_whs.shape)
```

After removing the outliers using zscore we losses 9 rows of data. Lets see our percent of our data contains that 9 rows lost

```python
# % of Data lost
print('Data Loss Percentage', ((WHS.shape[0]-new_whs.shape[0])/new_whs.shape[0])*100)
```

6% of our data lost is fine because is not up to 10%

#  Removing outliers using IQR

```python
from scipy import stats 
IQR = stats.iqr(WHS) 
IQR 
```

```python
Q1 = WHS.quantile(0.25)
Q3 = WHS.quantile(0.75)
```

```python
IQR = WHS[~((WHS < (Q1 - 1.5 * IQR)) |(WHS > (Q3 + 1.5 * IQR))).any(axis=1)]
IQR.shape
```

1 row was eliminated after using iqr method,  


```python
print('Data Loss Percentage', ((WHS.shape[0]-IQR.shape[0])/IQR.shape[0])*100)
```

Remeber  zscore elimited 9 rows which is 6% of our data lost but I will consider using zscore instead of using  iqr which eliminated 1 row which is 0.6% of data lost.However, since data lost in both methods are not up to 10% and to avoid seeing outliers again I will stick with zscore.

# Storing features in X and Label in Y

```python
x = df.drop(['Happiness Score'], axis =1)
y = df['Happiness Score']
```

# Model Building 

```python
# importing necessary libraries 
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.metrics import r2_score
```

# Spliting Data into Train and Test

```python
x_train, x_test,y_train, y_test = train_test_split(x, y,test_size=0.25,random_state =43)
```

# Linear Regression

```python
# Training model 
LRM = LinearRegression()
LRM.fit(x_train,y_train)
```

```python
print('LRM Train score:', LRM.score(x_train,y_train)*100) # checking model training score 
print('LRM Test Score:', LRM.score(x_test,y_test)*100) # checking model testing score 
LRM_pred = LRM.predict(x_test) # using model to predict on x_test data
print('\nMAE:',mean_absolute_error(y_test,LRM_pred))# Checking model performance and accuracy using MAE
print('MSE:', mean_squared_error(y_test,LRM_pred))# Checking model performance and accuracy using MSE
print('RMSE:', np.sqrt(mean_squared_error(y_test,LRM_pred)))# Checking model performance and accuracy using RMSE
print('R2 score:', r2_score(y_test,LRM_pred))# Checking model performance and accuracy using R2 SCORE

df_pred = pd.DataFrame({'Actual':y_test, 'Model Prediction':LRM_pred})#Creating DataFrame for Actual score and Model Prediction
df_pred.head(10)

```

```python
#Visualizing Actual score and prediction score 
plt.scatter(y_test,LRM_pred)
plt.xlabel('Actual Happiness Score')
plt.ylabel('Predicted Happiness Score')
plt.title('Actual vs Predicted')
plt.show()
```

# Support Vector Regression 

```python
# Training Model
svr = SVR()
svr.fit(x_train, y_train)
```

```python
print('SVR Train score:', svr.score(x_train,y_train)*100) # checking model training score 
print('SVR Test Score:', svr.score(x_test,y_test)*100) # checking model testing score 
svr_pred = svr.predict(x_test) # using model to predict on x_test data
print('\nMAE:',mean_absolute_error(y_test,svr_pred))# Checking model performance and accuracy using MAE
print('MSE:', mean_squared_error(y_test,svr_pred))# Checking model performance and accuracy using MSE
print('RMSE:', np.sqrt(mean_squared_error(y_test,svr_pred)))# Checking model performance and accuracy using RMSE
print('R2 score:', r2_score(y_test,svr_pred))# Checking model performance and accuracy using R2 SCORE

df1_pred = pd.DataFrame({'Actual':y_test, 'Model Prediction':svr_pred})#Creating DataFrame for Actual score and Model Prediction
df1_pred.head(10)

```

```python
#Visualizing Actual score and prediction score 
plt.scatter(y_test,svr_pred)
plt.xlabel('Actual Happiness Score')
plt.ylabel('Predicted Happiness Score')
plt.title('Actual vs Predicted')
plt.show()
```

# Random Forest Regression 

```python
# Training Model 
rdf = RandomForestRegressor()
rdf.fit(x_train,y_train)
```

```python
print('RDF Train score:', rdf.score(x_train,y_train)*100) # checking model training score 
print('SVR Test Score:', rdf.score(x_test,y_test)*100) # checking model testing score 
rdf_pred = rdf.predict(x_test) # using model to predict on x_test data
print('\nMAE:',mean_absolute_error(y_test,rdf_pred))# Checking model performance and accuracy using MAE
print('MSE:', mean_squared_error(y_test,rdf_pred))# Checking model performance and accuracy using MSE
print('RMSE:', np.sqrt(mean_squared_error(y_test,rdf_pred)))# Checking model performance and accuracy using RMSE
print('R2 score:', r2_score(y_test,rdf_pred))# Checking model performance and accuracy using R2 SCORE

df1_pred = pd.DataFrame({'Actual':y_test, 'Model Prediction':svr_pred})#Creating DataFrame for Actual score and Model Prediction
df1_pred.head(10)
```
